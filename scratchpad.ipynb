```python
import torch
import numpy as np
```

```python
x = np.array([2.0, 1.2, 4.5])
print(x)
print(type(x))
```

```python
t_x = torch.Tensor(x)
print(t_x)
print(type(t_x))
print(t_x.type())
```

```python
t_x32=torch.FloatTensor(np.float32(x))

print(t_x32)
print(type(t_x32))
print(t_x32.type())
```
